# Images reading texts

## Importing

```python
# read text from image
import easyocr
# showing image on notebook
from IPython.display import Image
```

## Using example

```python
# loading object
reader = easyocr.Reader(['pt'])
# Reading images
text_loaded = reader.readtext('/home/diegopmayer/Pictures/consistencia_2.png')

# showing the text readed
for text in text_loaded:
    print(f"Text readed:\n"
          f"{text[1]}")
    
#loaing image to view and compare
Image('/home/diegopmayer/Pictures/consistencia_2.png')
```
